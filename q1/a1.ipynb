```python
import numpy.matlib
import numpy as np
import math
```

```python
X = np.loadtxt('./data/x.dat')
```

```python
Y = np.loadtxt('./data/y.dat')[np.newaxis]
```

```python
Y = np.transpose(Y)
```

```python
def magsqr(x): 
    return sum(i**2 for i in x)
```

```python
def lwlr(x_train, y_train, x, tau):
    m = x_train.shape[0]
    n = x_train.shape[1]
    theta = np.zeros([2,1])
    lmbda = 0.0001
    w = np.zeros([m,1])
    for i, xi in enumerate(x_train):
        w[i] = np.exp(-magsqr(xi-x)/2/tau/tau)
    print(w)
    print("######################################")
    ##g is first derivative  
    g = np.ones([n,1])
    i = 1
    while(math.sqrt(magsqr(g)) > 1e-06):
        #logistic regression function htheta
        print(i)
        print("#########h############")
        h = 1 / (1+np.exp(-np.dot(x_train,theta)))
        print(h)
        print("#########g###########")
        g = np.dot(np.transpose(x_train),w*(y_train-h)) - lmbda*theta
        print(g)
        print("#########D#########")
        D = np.diag(np.transpose(-w*h*(1-h))[0])
        print(D)
        print("######Hessian#####")
        H = np.dot(np.dot(np.transpose(x_train),D),x_train) - np.diag(lmbda*np.ones(n))
        print(H)
        print("#######theta#######")
        theta -= np.dot(np.linalg.inv(H),g)
        print(theta)
        i+=1
       
    
    return (1 / (1+np.exp(-np.dot(x,theta))))[0]
```

```python
lwlr(X,Y,[1.6359447e-01, -9.2105263e-01],0.1)
```
